## Introduction

In this project, I will clean and explore two datasets. The first dataset comes from the FBI's National Instant Criminal Background Check System. The NICS is used to determine whether a prospective buyer is eligible to buy firearms or explosives. Gun shops call into this system to ensure that each customer does not have a criminal record or isn’t otherwise ineligible to make a purchase. The second dataset has been supplemented with state level data from census.gov.

The NICS data is found in one sheet of an .xlsx file. It contains the number of firearm checks by month, state, and type. The U.S. census data is found in a .csv file. It contains several variables at the state level. Most variables just have one data point per state (2016), but a few have data for more than one year.

By the end of this project, I will have addressed several questions: 
- How have guns sales changed over time?
- What are the top 5 state in guns per capita? And how has this changed over time?
- Which states have had the highest growth in guns sold per capita from 2010-2016?
- What census data is most associated with high gun per capita?

#### More information about the NICS data and calculating gun sales
The original PDF contains important notes and caveats, such as:

> These statistics represent the number of firearm background checks initiated through the NICS They do not represent the number of firearms sold Based on varying state laws and purchase scenarios, a one-to-one correlation cannot be made between a firearm background check and a firearm sale

To convert background checks into gun sales, I relied on the methodology used in this [New York Times analysis](http://www.nytimes.com/interactive/2015/12/10/us/gun-sales-terrorism-obama-restrictions.html) which was suggested in the [2013 Small Arms Survey](http://www.smallarmssurvey.org/fileadmin/docs/F-Working-papers/SAS-WP14-US-Firearms-Industry.pdf) by Jurgen Brauer, a professor at Georgia Regents University:

> The magnitude of the demand for firearms in the United States can be approximated if one is willing to make two assumptions: firstly, that all permit checks are routine procedural checks by states against FBI records and are not associated with an intent to purchase a gun; and, secondly, that all in-store (retailer) checks by licensed firearms dealers against FBI records result in at least one firearms purchase. In terms of these assumptions, the percentage of in-store checks out of all NICS checks yields an approximated annual demand. 

> As approximations go, one may then add ‘handgun’ checks, plus ‘long gun’ checks, plus two ‘multiple’ checks (at least one handgun and one long gun), and augment the resulting number by a factor of 1.1, termed here the multiple gun sales factor (MGSF). 

Permit checks and other types of checks were omitted because each state maintains its own rules regarding the frequency, if any, with which its issued permits are checked against FBI records. The multiplier is an estimate based on Mr. Brauer's interviews with gun shop owners.

### NICS Data Cleaning

Based on the methodology outlined in the [Small Arms Survey](http://www.smallarmssurvey.org/fileadmin/docs/F-Working-papers/SAS-WP14-US-Firearms-Industry.pdf) and the questions I proposed in the introduction, I will perform the following steps in order to clean the NICS and prepare it for further analysis:

1. Fill null values with 0.0 in order to perform calculations on the data
2. Creates four new columns, a handgun total column, long gun total column, a sales estimate column, and a year column
3. Remove all columns except for the 'month', 'state', 'multiple', and the previous created columns
5. Removes rows with 'state' values that are not found in the census data, which include the District of Columbia, Guam, Mariana Islands, Puerto Rico, and the Virgin Islands. Also removes rows with a year of 1998 or 2017. This is because the 1998 data appears incomplete, likely because it was the first year of the system and it had not been fully implemented. 2017 data was omitted because the last month in the dataset was from September. Note - these data sets were downloaded from the Google Doc in the project, not directly from NICS site.
6. Convert all columns after month and state to integers
7. Reset the indexes


```python
#replace NAs with 0s
df_nics.fillna(0.0,inplace=True)
```

```python
#create total column for all check on handguns
df_nics['handgun_totals'] = (df_nics.handgun + df_nics.prepawn_handgun + 
                             df_nics.redemption_handgun + df_nics.returned_handgun + df_nics.rentals_handgun + 
                             df_nics.private_sale_handgun + df_nics.return_to_seller_handgun)


#create total column for all check on long guns
df_nics['long_gun_totals'] = (df_nics.long_gun + df_nics.prepawn_long_gun + df_nics.redemption_long_gun + 
                              df_nics.returned_long_gun + df_nics.rentals_long_gun + df_nics.private_sale_long_gun + 
                              df_nics.return_to_seller_long_gun)

#create a new column based on the proposed methodology from the Small Arms Survey
MGSF = 1.1
df_nics['sales_estimate'] = MGSF * (df_nics.handgun_totals + df_nics.long_gun_totals + df_nics.multiple * 2)

#creates a year column from the month column
df_nics['year'] = df_nics['month'].apply(lambda x: x[:4])
```

**Note:** Based on the proposed methodology from the Small Arms Survey, we will also use the 'mutiple' column to calculated estimated gun sales. This is the only 'multiple' column so we will not need to create a separate multiple_total column

```python
#identifies the columns that will be kept
keep_cols = ['month', 'state', 'multiple', 'handgun_totals', 'long_gun_totals', 'sales_estimate', 'year']

#removes columns that are not in the above list to simplify the data we work with
#remove=[]
#for col in list(df_nics.columns):
    #if not col in keep_cols:
        #remove.append(col)
#df_nics.drop(remove, axis=1, inplace=True)
df_nics.filter(keep_cols)
```

```python
#function for removing rows with unnecessary data
def remove_rows(col_condition, list_filter):
    remove_vals = df_nics.query('{} {}'.format(col_condition, list_filter))
    df_nics.drop(remove_vals.index, inplace=True)
    return df_nics
```

```python
#removes information for 'states' that are not in the census data such as 'District of Columbia', 
#'Guam', 'Mariana Islands', 'Puerto Rico', and the 'Virgin Islands'
df_nics = remove_rows('not state in', list(df_census.columns[2:]))

#Removes all data from 1998 and 2017
df_nics = remove_rows('year in', ['1998', '2017'])
```

```python
#turns all columns into integers for consistency
for col in df_nics.columns[2:]:
    df_nics[col] = df_nics[col].astype(int)
```

```python
#resets the index for the leaner dataframe
df_nics.reset_index(inplace=True, drop=True)
```

Steps:

By removing the notes information in row 64 and below, the 3 duplicate rows are eliminated

```python
#transposes the census data so the states are in the index. This will make it easier to merge the two data sets 
df_census = df_census.transpose()

#turns the Fact row into the column names
headers = df_census.iloc[0,:]
df_census.columns = headers
```

```python
#removes the Fact and Fact Note rows
df_census.drop(['Fact', 'Fact Note'],axis=0,inplace=True)
```

```python
#replace Null values with 0
df_census.fillna("0",inplace=True)

#replaces the value flags in some of the cells with 0
val_flags = ['-','D','F','FN','NA','S','X','Z'] #value flags found in bottom of data.
df_census.replace(to_replace=val_flags, value="0", inplace=True) #replace all value flags with 0
```

```python
#curates values so they can be converted from strings to floats
for col in df_census.columns:
    #first, if a value in a column has  a % sign, remove it and divide it by 100, else remove all commas and $
    df_census[col] = df_census[col].apply(lambda x: x.replace(',','').replace('$','').replace('%',''))
    df_census[col] = df_census[col].astype(float) #turn all state columns to floats
```

```python
pop2016 = 'Population estimates, July 1, 2016,  (V2016)'
#creates two columns, one for employment percentage and 
#the other for the percentage of the state population that is a veteran

df_census['employment%_2015'] = df_census['Total employment, 2015'] / df_census[pop2016] * 100
df_census['veterans%_2015'] = df_census['Veterans, 2011-2015'] / df_census[pop2016] * 100
```

### Merge the two datasets

```python
#Groups the series in the NICS data by year and state, sums the values for each year and state
nics_year_state = df_nics.groupby(['year','state'], as_index=False, sort=False).sum()
```

```python
#function for merging the two data sets by state, based on the year in the NICS data.
def create_merge(year):
    
    merged = nics_year_state.query('year == "{}"'.format(year)).merge(df_census, how="inner", left_on='state', right_index=True).drop('year',axis=1)
    merged['sold_per_capita'] = merged.sales_estimate / merged['population_{}'.format(year)] * 100
    merged.sort_values('sold_per_capita', ascending=False, inplace=True)
    merged.reset_index(drop=True, inplace=True)
    
    return merged
```

```python
#Because the census data does not have information for each year like the NICS data, we can only examine 2016 and 2010
merged_2016 = create_merge('2016')
merged_2010 = create_merge('2010')
```

```python
#Creates a column in merged_2016 that shows the % change in sales per capita from 2010-2016
merged_2016['sales_chg_10to16'] = (merged_2016.sales_estimate - merged_2010.sales_estimate) / merged_2016.population_2010 * 100
```

### How have US guns sales changed over time?

```python
#Groups the series in the NICS data by year, sums the values for each year
nics_year= df_nics.groupby(['year'], as_index=False).sum()
```

```python
#plot of years on x axis and sales estimates on y (divided by 1mil for formatting)
plt.figure(figsize=(15,5))
plt.plot(nics_year.year, nics_year.sales_estimate/1000000) #estimates in millions

#Chart formatting
plt.title('U.S. Gun Sales from 1999-2016', fontsize=16)
plt.ylabel('Estimated Guns Sold (Millions)', fontsize=14)
plt.xticks(nics_year.year)
plt.xlabel('Year', fontsize=14);

yrbeg = nics_year.sales_estimate[0]
yrend = nics_year.sales_estimate[17]
growth = (yrend / yrbeg)**(1/(2016-1999)) - 1

print("1999 Gun Sales: {}".format(yrbeg), 
      "2016 Gun Sales: {}".format(yrend), 
      "Difference: {}".format(yrend - yrbeg), 
      "Average Annual Growth Rate: {:.2f}%".format(growth*100),
     sep="\n")
```

### What census data is most associated with high gun per capita? 

```python
correlation = merged_2016.corr(method='pearson')['sold_per_capita'].sort_values(ascending=False)
top_corr_abs = merged_2016.corr(method='pearson')['sold_per_capita'].abs().sort_values(ascending=False)[1:5]
```

```python
def scatter_func(col1, col2):
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))
    
    #Left chart
    ax1.scatter(merged_2016.sold_per_capita, merged_2016[col1])
    #Left chart formatting
    ax1.set_title('Guns Sold per Capita vs {}'.format(col1[:36].title()), fontsize=13)
    ax1.set_xlabel('Guns Sold per Capita (%)')
    ax1.set_ylabel(col1.title())
    
    #Right chart
    ax2.scatter(merged_2016.sold_per_capita, merged_2016[col2])
    #Right chart formatting
    ax2.set_title('Guns Sold per Capita vs {}'.format(col2[:36].title()), fontsize=13)
    ax2.set_xlabel('Guns Sold per Capita (%)')
    ax2.set_ylabel(col2.title());
```

```python
i = 0
while i < len(top_corr_abs.index):
    scatter_func(top_corr_abs.index[i], top_corr_abs.index[i+1])
    i+=2
    
print("Top Correlated Census Data")
for val in top_corr_abs.index:
    print("{}:  {}".format(val,correlation[val]))
```

**What census data is most associated with high gun per capita?**
-	The data that appears to have the highest correlation with a high number of guns sold per capita were a larger percentage of veterans in the state's population, a less dense state in terms of population per square mile, cheaper gross rents, and cheaper homeowner costs. Of the top 4 correlated census data, all had a correlation between 0.5 and 0.6
-	_Higher Percentage of Veterans:_ Since veterans are familiar with firearms, it isn’t unsurprising that the higher the number of veterans the higher the guns sold per capita. However, since the percentage of veterans for all states is below 10%, this is unlikely to be the contributing factor to higher gun sales. Most likely, veterans are attracted to live in states where more guns are sold. The data does not provide enough information to confirm if more guns alone or regulations supporting more gun sales are an attractive attribute for veterans.
-	_More rural states:_ The smaller the population per square mile, the higher the guns sold per capita. This make logical since we know the more rural areas are more likely to have more guns for hunting or other purposes. States with a larger rural population are also more likely to have more lenient gun laws.
-	_Cheaper rents and ownership costs:_ Selected monthly owner costs are the sum of payments for mortgages, deeds of trust, contracts to purchase, or similar debts on the property; real estate taxes; fire, hazard, and flood insurance on the property; utilities; and fuels. It also includes, where appropriate, the monthly condominium fee for condominiums and mobile home costs ([Source](https://www.census.gov/quickfacts/fact/note/US/HSG651217)). We see that the negative correlation with gross rent and Median Selected Monthly Owner Costs is understandable since less populated states and rural areas have less housing demand and therefore lower overall prices, be that rent prices or ownership prices.
